```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000
random_state = 2020
centers = [[0.25, 0.75]]
x, y = make_blobs(n_samples=n, centers=centers, random_state=random_state)
x_norm = (x - x.min(axis=0)) / x.ptp(axis=0)
# x_norm = x
```

```python
plt.figure()
plt.scatter(x_norm[:, 0], x_norm[:, 1], marker='.', c=y, s=25, edgecolor='face')
plt.grid(False)
# plt.xlim(0, 1.0)
# plt.ylim(0, 1.0)
plt.show()
```
